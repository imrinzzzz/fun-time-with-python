# Lab 1: setting up python
Python is an interpreted language. There are two basic mode for using Python.
1. **normal mode (commonly used)** <br>
    which is about creating script file “.py” where Python statements are written for a specific task and run this file in Python interpreter.
2.  **Python interactive mode** <br>
    which is a command line shell where each of input statement will be executed and gives immediate feedback after submission
    
### Python Interactive Shell
You should set up a PATH first. To open the shell, press `windows + R`, type `cmd` then type `python` and hit return. Or if you haven't set up a path, press `windows + R`, type `cmd` then type `cd Anaconda3` and type `python` and hit return. <br>
The display should look like this
```python
Python 3.6.5 |Anaconda, Inc.| (default, Mar 29 2018, 13:32:41)
Type "help", "copyright", "credits" or "license" for more information.
>>>
```
Once the Python interactive shell is started, you can issue any command at the command prompt `>>>`. To quit the python shell, you can type `exit()` or Press `(Ctrl+Z)`.

#### Numbers
```python
>>> print(5);\
>>> print(1.55);\
>>> print(True);
```
Every element in Python has it own type that determine the operation supported by the element. <br>
- bool&emsp;True or False.
- int&emsp;&emsp;Whole number value such as 3.4141 or -2.234.
- float&emsp; Numeric value with the decimal point such as 587 or -45.
- str&emsp;&emsp;String of characters inside the quote (usually double quote is used)

```python
>>> x = "100";\
>>> y = "250";\
>>> print(x+y);
>>> print(x, y);
```
The result is `100250` and `100 250` <br>

#### Read and Write Files <br>

The basic concept of file writing is dealing with file object by using the **open**, **read**, **write**, and close methods.
```python
# How to write file
>>> file = open("../Desktop/testfile.txt", "w");
>>> file.write("Hello, This is my file");
>>> file.write("\n--- nextline ---");
>>> file.close();

# How to read file
>>> file = open("../Desktop/testfile.txt", "r");
>>> read_1 = file.read();
>>> read_2 = file.read(5);
>>> read_3 = file.readline();
>>> read_4 = file.readline();
>>> file.close();
# use print() command to display all variable named read 1, read 2, read 3, and read 4

# How to append file
>>> file = open("../Desktop/testfile.txt", "a");
>>> file.write("\n -- append --");
>>> file.close();
```
