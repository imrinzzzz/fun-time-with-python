## Matplotlib
Matplotlib is one of the most popular Pythonâ€™s data visualization library. The
library was originally created to offer plotting feature in Python to work similarly
to plotting feature in MatLab, another programming language. Matplotlib
support visualizations for both 2D and 3D.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(1,3,'bo')
plt.plot(2,2,'ro')
```

```python
plt.plot([1,2,12,30])   # only tell y-axis
plt.plot([1,4,3,12])    # only tell y-axis
```

```python
plt.plot([0, 3, 5, 6], [0, 5, 15, 10])   # tells x-axis then y-axis
plt.plot([0, 1, 2, 6], [10, 8, 1, 3])    # tells x-axis then y-axis
```

```python
plt.plot([0, 3, 5, 6], [0, 5, 15, 10], '--',label='A',  # tells what kind of line, and label
color="pink", linewidth=5)                              # tells colour, and width
plt.plot([0, 1, 2, 6], [10, 8, 1, 3], ':', label='B',
color="lightblue", linewidth=2)
plt.legend()    # plot the mini square telling what line is A, B on the topleft of the graph
```

Note (â€™-â€™ solid line style, â€™â€“â€™ dashed line style, â€™-.â€™ dash-dot line style, â€™:â€™ dotted
line style)

```python
fig = plt.figure(figsize=(6,2))     # the size of the graph box
plt.plot([0, 3, 5, 6], [0, 5, 15, 10], label='A')
plt.plot([0, 1, 2, 6], [10, 8, 1, 3], label='B')
plt.legend()
```

```python
plt.barh([1,2,3],[0,1,5])   # bar graph
```

```python
import numpy as np
```

```python
n = 6
x = np.arange(n)
y1 = (1-x/float(n))
y2 = (1-x/float(n))
plt.bar(x, +y1, facecolor='blue', edgecolor='white')  # +y means above
plt.bar(x, -y2, facecolor='red', edgecolor='white')   # -y means below
```

```python
plt.barh([1,2,3],[0,1,5])
plt.barh([2,2,5],[1,1,1])
plt.bar([1,2,3],[3,4,5])
```

You should see the overlapping problem in the previous exercise.
You can solve it using **subplot**.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.barh([1,2,3],[0,1,5])
```

The digits in the number â€™111â€™ refers to The **three arguments**: 
- the number of rows
- the number of columns
- the plot number

Then, you can add more subplots.

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)
ax1.barh([1,2,3],[0,1,5])
ax2.barh([2,2,5],[1,1,1])
ax3.bar([1,2,3],[3,4,5])
```

```python
#  Scatter plot
fig = plt.figure()
ax = fig.add_subplot(111)
N = 30
x = np.random.rand(N)
y = np.random.rand(N)
scat = ax.scatter(x,y,200,edgecolors = 'k', facecolors = 'c')
```

```python
rng = np.arange(50)
rnd = np.random.randint(0,10,size=(3,rng.size))
yrs = 1950 + rng
fig, ax = plt.subplots(figsize=(5,3))

ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])  # set year(x-axis), numbers (y-axis), label name
ax.set_title('Combined debt growth over time')  # set the graph name
ax.legend(loc='upper left')   # set the location of legend
ax.set_ylabel('Total debt')   # set the label name on y-axis

ax.set_xlim(left=yrs[0], right=yrs[-1])  # or ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
```

### Seaborn
Seaborn is a library that is an extension of Matplotlib. The library **illustrates
related statistical information of a graph** to allow us to gain more insight about
a dataset. Seaborn also closely integrated with pandas data structures. Seaborn
offers many visualization tools

make sure you use the lastest seaborn.
`>>> conda update seaborn`

```python
import seaborn as sns
```

```python
sns.set()
tips = sns.load_dataset("tips")   # load from external file
sns.relplot(x="total_bill", y="tip", hue="smoker", style = "smoker",
          size="size", data=tips);
```

```python
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="time", hue="smoker",
            style="smoker", size="size", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", col="time", hue="size",
            style="smoker", size="size", palette="ch:r=-.5,l=.75", markers=["D","o"], 
            sizes=(10, 125), edgecolor=".2", linewidth=.5, alpha=.75,
            data=tips);
```

The attribute `kind` changes the type of plotting. We'll use
- line
- swarm
- violen
- bar

```python
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="smoker", hue="sex",
            kind="line", data=tips);
```

```python
import matplotlib.pyplot as plt
sns.set()
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", hue="smoker", kind="swarm", 
           data=tips);
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="violin", split=True, data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="bar",data=tips);
```
